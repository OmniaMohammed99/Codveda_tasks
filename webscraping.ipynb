```python
import requests
from bs4 import BeautifulSoup as soup
import csv
```

```python
source = requests.get('https://faq.usps.com/s/article/2025-Postage-Price-Change')
```

```python
webpage = soup(source.content,features= 'html.parser')
```

```python
#print out html
print(webpage.prettify())
```

```python
product = []
current_prices = []
planned_prices = []
```

```python
for i in webpage.find_all(class_='postage_price_table'):

  string = i.text #find and find all returns whole tag
  product.append(string.strip()) #strip to clear spaces


for i in webpage.find_all(class_='postage_price_table'):
  string = i.text
  substring = string.rpartition('$')[2]
  current_prices.append(substring.strip())


for i in webpage.find_all(class_='postage_price_table'):
  string = i.text
  substring = string.rpartition('$')[2]
  planned_prices.append(substring.strip())
```

```python
file_name = 'postageServicesrPrices.csv' #name of the file

with open(file_name, 'w', encoding = 'utf-8') as f:
  f.write = csv.writer(f)
  f.write.writerow(['product','current_prices','planned_prices']) #headers of the file

  for i in range(len(product)):
    f.write.writerrow([product[i],current_prices[i],planned_prices[i]])
```
