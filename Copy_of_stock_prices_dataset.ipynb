#Feature engineering

```python
# creating new feature(returns)
df['daily_return'] = (df['close'] - df['open']) / df['open']
```

```python
# creating new feature(volatility)
df['daily_volatility'] = (df['high'] - df['low']) / df['open']
```

```python
#drop unnumerical data
data = df.drop(columns='symbol')
data.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series into trend, seasonality, and residuals
decomposition = seasonal_decompose(df['close'], model='additive', period=365) # period=365 for yearly seasonality

# Plot the decomposed components
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(decomposition.observed, label='Observed')
plt.legend(loc='best')
plt.title('Observed Time Series')

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='best')
plt.title('Trend Component')

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality')
plt.legend(loc='best')
plt.title('Seasonal Component')

plt.subplot(414)
plt.plot(decomposition.resid, label='Residuals')
plt.legend(loc='best')
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

data preperation

```python
#split the data
x = data.drop(columns='close')
y = data['close']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

scaling the data


```python
from sklearn.preprocessing import MinMaxScaler

# Scale features (X)
x_scaler = MinMaxScaler()
x_train_scaled = x_scaler.fit_transform(x_train)
x_test_scaled = x_scaler.transform(x_test)

# Scale target (y)
y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))
```

```python
y_test_scaled
```

#Modeling

```python
# Build LSTM model
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
model = keras.models.Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(x_train_scaled.shape[1], 1)))
model.add(LSTM(64))
model.add(Dense(128, activation ='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss='mae',metrics=[keras.metrics.RootMeanSquaredError()], optimizer='adam')
```

```python
training = model.fit(x_train_scaled, y_train_scaled, epochs=20, batch_size=32, validation_split=0.1)
```

```python
prediction = model.predict(x_test_scaled)
```

```python
model_loss = model.evaluate(x_test_scaled,y_test_scaled)
print('Model loss is ',model_loss)
```

```python
#plot the prediction
plt.figure(figsize=(12, 6))
plt.plot(y_test_scaled, label='Actual')
plt.plot(prediction, label='Predicted')
plt.legend()
```
